```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Flatten, Dense, Dropout, Lambda
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.python.keras.utils.vis_utils import plot_model
from keras import backend as K
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageFont, ImageDraw
import random
```

```python
def create_pairs(x, digit_indices):
    '''
    Create positive and negative pairs of images to compare.
    '''
    
    pairs = []
    labels = []
    n = min([len(digit_indices[d]) for d in range(10)]) - 1
    for d in range(10):
        for i in range(n):
            z1, z2 = digit_indices[d][i], digit_indices[d][i+1]
            pairs += [[x[z1], x[z2]]]
            inc = random.randrange(1, 10)
            dn = ( d + inc ) % 10
            z1, z2 = digit_indices[d][i], digit_indices[dn][i]
            pairs += [[x[z1], x[z2]]]
            labels += [1, 0]
    return np.array(pairs), np.array(labels)
```

```python
def create_pairs_on_set(images, labels):
    digit_indices = [np.where(labels ==i)[0] for i in range(10)]
    pairs, y = create_pairs(images, digit_indices)
    y = y.astype('float32')
    return pairs, y
```

```python
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

```python
train_images = train_images.astype('float32')
test_images = test_images.astype('float32')
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
tr_pairs, tr_y = create_pairs_on_set(train_images, train_labels)
ts_pair, ts_y = create_pairs_on_set(test_images, test_labels)
```

```python
def show_image(image):
    plt.figure()
    plt.imshow(image)
    plt.colorbar()
    plt.grid(False)
    plt.show()
```

# Create BASE NETWORK

```python
def initialize_base_network():
    input_layer = Input(shape=(28,28,), name="base_input")
    x = Flatten(name="flatten_input")(input_layer)
    x = Dense(128, activation="relu", name="first_base_dense")(x)
    x = Dropout(0.1, name="first_dropout")(x)
    x = Dense(128, activation="relu", name="second_base_dense")(x)
    x = Dropout(0.1, name="second_dropout")(x)
    x = Dense(128, activation="relu", name="third_base_dense")(x)
    return Model(inputs=input_layer, outputs=x)
```

```python
def euclidean_distance(vects):
    x, y = vects
    sum_square = K.sum(K.square(x-y), axis=1, keepdims=True)
    return K.sqrt(K.maximum(sum_square, K.epsilon()))

def eucl_dist_output_shape(shapes):
    shape1, shape2 = shapes
    return (shape1[0], 1)
```

```python
base_network = initialize_base_network()
```

```python
plot_model(base_network, show_shapes=True, show_layer_names=True, to_file="siamese-base-model.png")
```

# Functional API to create TWO different inputs onto the base-network

```python
input_a = Input(shape=(28, 28,), name="left_input")
vect_output_a = base_network(input_a)
input_b = Input(shape=(28, 28,), name="right_input")
vect_output_b = base_network(input_b)
```

```python
output = Lambda(euclidean_distance, 
                name="output_layer", 
                output_shape=eucl_dist_output_shape)([vect_output_a, vect_output_b])
```

```python
model = Model([input_a, input_b], output)
```

```python
plot_model(model, show_shapes=True, show_layer_names=True, to_file="siamese-outer-model.png")
```
